```python
#1.two plots displaying the distribution of a single continuous variable
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

#histogram for bedrooms
sns.histplot(df['bedrooms'], bins=15, ax=ax[0], color='purple', kde=True)
ax[0].set_title('Histogram of Number of Bedrooms')
ax[0].set_xlabel('Number of Bedrooms')
ax[0].set_ylabel('Count')

#boxplot for bedrooms
sns.boxplot(x=df['bedrooms'], ax=ax[1], color='purple')
ax[1].set_title('Boxplot of Number of Bedrooms')
ax[1].set_xlabel('Number of Bedrooms')
ax[1].set_ylabel('Count')
plt.show()
```

```python
#2  two plots displaying information about a single categorical variable
#A countplot for the 'title' column
plt.figure(figsize=(10, 7))
sns.countplot(data=df, y='title', order=df['title'].value_counts().index, palette='pastel')
plt.title('Count Plot of House Types')
plt.xlabel('Count')
plt.ylabel('House Type')
plt.show()
```

```python
#A violin plot for the 'title' column
plt.figure(figsize=(12, 8))
sns.violinplot(data=df, y='title', x='bedrooms', palette='pastel')
plt.title('Violin Plot of House Types vs. Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('House Type')
plt.show()
```

```python
#3. one plot displaying information about both a continuous variable and a categorical variable
plt.figure(figsize=(14, 8))
#A bar plot showing average price for each type of house
sns.barplot(x="title", y="price", data=df, errorbar=None, palette="coolwarm")
plt.title("Average House Prices by Type of House", fontsize=16)
plt.xlabel("Type of House", fontsize=14)
plt.ylabel("Average Price (in NGN)", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#4. two plots should display information that shows a relationship between two variables
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
# Scatter plot of bedrooms vs price
axes[0].scatter(df['bedrooms'], df['price'], alpha=0.8, color='purple')
axes[0].set_title('Price vs. Number of Bedrooms')
axes[0].set_xlabel('Number of Bedrooms')
axes[0].set_ylabel('Price (in billions)')
axes[0].set_yscale('log')
# Scatter plot of bathrooms vs price
axes[1].scatter(df['bathrooms'], df['price'], alpha=0.8, color='blue')
axes[1].set_title('Price vs. Number of Bathrooms')
axes[1].set_xlabel('Number of Bathrooms')
axes[1].set_ylabel('Price (in billions)')
axes[1].set_yscale('log')

plt.show()
```

```python
#5.one plot displaying the information about 3 or more variables
#correlation matrix
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, square=True)
plt.title('Correlation Heatmap', size=16)
plt.show()
```
